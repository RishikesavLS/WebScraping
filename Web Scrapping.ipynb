### Import Library

```python
from bs4 import BeautifulSoup
```

```python
import requests
```

### Setting scrapping website

```python
url = "https://m.imdb.com/title/tt0372784/fullcredits/cast"
```

```python
allow = requests.get(url)
```

```python
print(allow)
```

### Testing Scrapping Allowability

```python
soup = BeautifulSoup(allow.text)
```

```python
soup
```

### Scrapping Cast List of "Batman Begins"

```python
cast = soup.select("h4")
```

```python
cast
```

```python
actor_name = [] # to store items in cast
```

```python
for item in cast: 
    actor_name.append(item.text)
```

```python
actor_name
```

```python
characters = soup.select("#fullcredits-content .unbold")
```

```python
characters
```

```python
role_name = [] # To store items in characters
```

```python
for item in characters:
    role_name.append(item.text)
```

```python
role_name
```

```python
role_edit = [] # To store items of role_name from which "\n" character is removed
```

```python
for item in role_name:
        role_edit.append(item.replace("\n",""))
```

```python
role_edit
```

```python
role_final = [] # To store items of role_name from which "(uncredited)" is removed
```

```python
for item in role_edit:
    role_final.append(item.replace("(uncredited)",""))
```

```python
role_final
```

### Creating Dataframe

```python
import pandas as pd
```

```python
credits = pd.DataFrame(role_final,columns=['Character'])
```

```python
credits
```

```python
credits['Actor'] = actor_name
```

```python
credits
```

### Saving as .CSV

```python
credits.to_csv('credit.csv')
```
